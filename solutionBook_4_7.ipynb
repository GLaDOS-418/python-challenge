# challenge 4 : follow the chain

**_url_** : http://www.pythonchallenge.com/pc/def/linkedlist.php<br/>
**_solution_** : the image has a href to <http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing=12345>.<br/>
>On visiting the link, there is a one liner text giving the value of next nothing. And a chain follows...

It is an attempt to introduce the _urllib.request_ module of python. Send the GET requests and keep following the chain until you can't.<br/>
I'm using **_httplib2_** though. It's a third party library for handling http requests better than python's standard libraries. What exactly? find [here](http://www.diveintopython3.net/http-web-services.html)

A **digression** indeed but, rather an interesting one:
[why-am-i-getting-an-unboundlocalerror-when-the-variable-has-a-value](https://docs.python.org/3/faq/programming.html#why-am-i-getting-an-unboundlocalerror-when-the-variable-has-a-value)

```python
import httplib2
import re
```

```python
hr = httplib2.Http('.cache')
totalRequests = 0

def keepTrackingUntil(url, nothing, total):
    response, content = hr.request(url+nothing)
    content = content.decode('utf-8')
    print(content)
    
    p = re.compile(r'\d+$') #find next nothing
    next_nothing = p.search(content)
    if next_nothing != None:
        return keepTrackingUntil(url, next_nothing[0], total+1)
        pass
    return total
    pass

totalRequests = keepTrackingUntil('http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing=','12345', 0) #seed one.
print('total requests till now: ' + str(totalRequests))
```

```python
totalRequests = keepTrackingUntil('http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing=','8022', totalRequests)
print('total requests till now: ' + str(totalRequests))
```

```python
import re
import zipfile as z
```

```python
p = re.compile(r'\b\d+$')

count =0
num_list = []
def keepCheckingUntil(filename):
    global count, num_list
    count+=1
    with open('channel/{0}.txt'.format(filename), 'r') as f:
        x = "".join([ line for line in f])
        res = re.search(p, x)
        print(x)
        if res != None:
            num_list.append(int(res[0]))
            keepCheckingUntil(res[0])
            pass
        pass
    pass

keepCheckingUntil('90052')
print(count)
```

```python

with z.ZipFile('channel.zip', 'r') as myzip:
    hint = []
    fileName='90052'
    while fileName!=None:
        with myzip.open('{0}.txt'.format(fileName)) as f:
            #print(myzip.infolist())
            hint.append(myzip.getinfo('{0}.txt'.format(fileName)).comment.decode('utf-8'))
            res = re.search(p, f.read().decode('utf-8'))
            fileName = res[0] if res != None else None
            pass
        pass
    hint = "".join(hint)
    print(hint)    
```
